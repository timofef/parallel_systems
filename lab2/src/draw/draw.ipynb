```python
import matplotlib.pyplot as plt
```

```python
x = [2, 4, 8, 16, 32, 64]
```

```python
plt.plot(x, [1.713584, 3.197999, 6.186849, 12.170662, 24.118778, 47.951857], '-', label="$S_{1}(N), C_{f}=0.5*10^{7}$")
plt.plot(x, [1.646893, 3.011558, 5.588416, 10.483679, 18.735854, 32.277214], '-.', label="$S_{1}(N), C_{f}=5*10^{4}$")
plt.xlabel("N")
plt.ylabel("S(N)")
plt.grid()
plt.xlim(0)
plt.ylim(0, 65)
plt.plot(x, x, ':', label="S(N)=N")
plt.legend()
plt.show()
```
